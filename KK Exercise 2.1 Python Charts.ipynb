```python
# Import libraries

import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Read data source
    
wp_df = pd.read_excel("ex2-2/world-population.xlsm")

# display firt 5 records
wp_df.head(5)
```

```python
# Plot the line chart
  
plt.plot(wp_df['Year'], wp_df['Population'], color='blue', marker='o')
plt.title('Population Vs Year', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Population', fontsize=14)
plt.grid(True)
plt.show()
```

```python
# Plot the step chart
  
plt.step(wp_df['Year'], wp_df['Population'], 'g^', where='post')
plt.title('Population Vs Year', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Population', fontsize=14)
plt.grid(True)
plt.show()
```
