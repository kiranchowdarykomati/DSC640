```python
# Load necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data into Data Frames
crimerate_DF = pd.read_csv("ex4-2\\crimerates-by-state-2005.csv")
birthrate_DF = pd.read_csv('ex4-2\\birth-rate.csv')

# Remove United States as it is a country
crimerate_DF = crimerate_DF[crimerate_DF.state != 'United States']
```

```python
# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(10, 7))

# Scatter Plot
plt.plot('murder', 'burglary', data=crimerate_DF, linestyle='none', markersize=10, 
         markerfacecolor='black', marker='o', markeredgecolor="none")

# Add chart title and labels
plt.title("Murders vs Burglaries in the US(Scatter Plot)", fontsize = 14, loc = 'left')
plt.xlabel("Murders", fontsize = 11)
plt.ylabel("Burglaries", fontsize = 11)

plt.show()
```

```python
# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(10, 6))

# Bubble Chart
sc = plt.scatter('murder', 'burglary', s=crimerate_DF.population*.0001, data=crimerate_DF, 
                 alpha=0.7, color='brown', label='Population')

# Add chart title and labels
plt.title("Murders vs Burglaries in the US(Bubble Chart)", fontsize = 15, loc = 'left')
plt.xlabel("Murders", fontsize = 12)
plt.ylabel("Burglaries", fontsize = 12)

# Add legend for population sizes                   
handles, labels = sc.legend_elements(prop="sizes", alpha=0.4, num=3)     
labels = ["10 MM", "20 MM", "30 MM"]     
legend = ax.legend(handles, labels, loc="lower right", title="Population",fontsize=11, title_fontsize=14) 

plt.show()
```

```python
# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(10, 6))

# Density plot
sns.kdeplot(birthrate_DF['1980'], color='black', fill='black')

# Add chart title and labels
plt.title("World Distribution of Birth Rates in the year 1980", fontsize = 15, loc = 'left')
plt.xlabel("Live Births per 1,000 Population", fontsize = 12)
plt.ylabel("Density", fontsize = 12)
```
