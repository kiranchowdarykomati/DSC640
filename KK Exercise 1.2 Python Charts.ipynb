```python
# Import libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Read data source
    
hcw_df = pd.read_excel("hotdog-contest-winners.xlsm")
hp_df = pd.read_excel('hotdog-places.xlsm')
oar_df = pd.read_excel('obama-approval-ratings.xls')

# display firt 5 records
hcw_df.head(5)
```

```python
plt.figure(figsize=(25,15))

colors = {'United States':'blue', 'Mexico':'green', 'Germany':'yellow', 'Japan':'red'}
c = hcw_df['Country'].apply(lambda x: colors[x])
plt.bar(hcw_df['Year'], hcw_df['Dogs_eaten'], color=c)
plt.ylabel('No. of Hot-Dogs', fontsize=20)
subtitle_string = 'Hot-Dogs Contest Winners'
title_string = 'Python Bar Chart'
plt.xlabel('Year', fontsize=20)
plt.suptitle(title_string, y=0.95, fontsize=20)
plt.title(subtitle_string, fontsize=20)
labels = list(colors.keys())
handles = [plt.Rectangle((0,0),1,1, color=colors[label]) for label in labels]
plt.legend(handles, labels)
plt.grid(True)
plt.show()
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(15,10))
plt.rcParams["figure.autolayout"] = True

plt.ylabel('Issues', fontsize=25)
plt.xlabel('Response in Percentage', fontsize=25)
subtitle_string = 'Performance of President in Last Quarter'
title_string = 'Stacked Bar Chart'
plt.suptitle(title_string, y=0.95, fontsize=25)
plt.title(subtitle_string, fontsize=25)
b1 = plt.barh(oar_df['Issue'], oar_df['Approve'],color="green")
b2 = plt.barh(oar_df['Issue'], oar_df['Disapprove'],left=oar_df['Approve'], color="red")
b3 = plt.barh(oar_df['Issue'], oar_df['None'], left=oar_df['Approve'] + oar_df['Disapprove'],color="orange")
plt.legend([b1, b2, b3], ["Approve", "Disapprove", "None"], title="Survey␣Rating", loc="upper right", bbox_to_anchor=([0.95, 1, .05, 0.12]))
plt.show()
```

```python
# Set up index of country labels, sorted by number of instances

countryLabels = hcw_df.groupby('Country')['Year'].nunique().sort_values(ascending=False).index

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(7, 7))

hcw_df['Country'].value_counts()

colors = sns.color_palette('colorblind')
# Create pie chart
plt.pie(hcw_df['Country'].value_counts(), autopct='%.0f%%', colors=colors, labels=countryLabels)

# Set up chart title
plt.title("Hotdog Eating Competition: Wins by Country", fontsize=15)
plt.show()
```

```python
# Initialize the matplotlib figure

f, ax = plt.subplots(figsize=(7, 7))

# Create pie chart
plt.pie(hcw_df['Country'].value_counts(), autopct='%.0f%%', labels=countryLabels, colors=colors, pctdistance=0.85)
plt.show

# Draw circle
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
# Adding Circle in Pie chart
fig.gca().add_artist(centre_circle)

# Set up chart title
plt.title("Hotdog Eating Competition: Wins by Country", fontsize=15)
plt.show()
```
